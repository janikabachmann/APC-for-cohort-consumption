```python
#This is final model: head of 2+ household 30-34 years old in 1984
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
data=pd.read_excel('hello/cohort11.xlsx',sheet_name='Sheet1')
df=pd.DataFrame(data)
print(df.head(2))
```

```python
ax=sns.regplot(x='unemployment change',y='11consumptionchange',data=df)
```

```python
ax=sns.regplot(x='maleregularchange',y='11consumptionchange',data=df)
```

```python
X=df[['11annualincomechange','unemployment change','maleregularchange']]
Y=df['11consumptionchange']
X=sm.add_constant(X)
model=sm.OLS(Y,X).fit()
predictions=model.predict(X)
print_model=model.summary()
print(print_model)
print(model.params)
```
